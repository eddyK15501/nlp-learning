```python
from tensorflow.keras.preprocessing.text import one_hot
```

```python
sentences = [
    'Aliens stole my homework yesterday',
    'Giraffes are terrible at limbo',
    'My pillow speaks French',
    'Zombies hate morning yoga',
    'Monkeys won the lottery',
    'Cats secretly rule everything',
    'Bananas fear the blender'
]
```

```python
# Define vocabulary size
voc_size = 10000
```

```python
# Integer encoding of words using the one_hot function
one_hot_repr = [one_hot(words, voc_size) for words in sentences]

one_hot_repr
    
```

```python
# Word embeddings
from tensorflow.keras.layers import Embedding, Input
from tensorflow.keras.models import Sequential
from tensorflow.keras.utils import pad_sequences
import numpy as np
```

```python
# Ensure sentence lengths are equal
max_sentence_length = 8

# 'pre' adds zeros at the front of the sentence vector
# 'post' adds at the end
embedded_docs = pad_sequences(one_hot_repr, padding='post', maxlen=max_sentence_length)

print(embedded_docs)
```

```python
# Feature representation
dimensions = 10
```

```python
model = Sequential()
model.add(Input(shape=(max_sentence_length,)))
model.add(Embedding(input_dim=voc_size, output_dim=dimensions)) # OR model.add(Embedding(voc_size, dimensions))

model.compile('adam', 'mse') # OR model.compile(optimizer='adam', loss='mse')
```

```python
model.summary()
```

```python
model.predict(embedded_docs)
```
